Python Script to use Regression Model and Suggest Optimal Pricing Based on Input Parameters

```python
# Script to suggest optimal pricing based on input parameters
def suggest_price(location, listing_quality, reviews, property_type, rating, bedrooms, beds, baths, season):

    input_data = pd.DataFrame({
        'latitude': [location[0]],
        'longitude': [location[1]],
        'minimum_nights': [listing_quality['minimum_nights']],
        'number_of_reviews': [reviews['number_of_reviews']],
        'reviews_per_month': [reviews['reviews_per_month']],
        'availability_365': [listing_quality['availability_365']],
        'rating': [rating],
        'bedrooms': [bedrooms],
        'beds': [beds],
        'baths': [baths],
        'room_type_Entire home/apt': [1 if property_type == 'Entire home/apt' else 0],
        'room_type_Hotel room': [1 if property_type == 'Hotel room' else 0],
        'room_type_Private room': [1 if property_type == 'Private room' else 0],
        'room_type_Shared room': [1 if property_type == 'Shared room' else 0],
        'season_summer': [1 if season == 'summer' else 0],
        'season_winter': [1 if season == 'winter' else 0]
    })

    # One-hot encode input data
    input_data = pd.get_dummies(input_data)

    # Ensure all columns match the training data
    for col in X_train.columns:
        if col not in input_data.columns:
            input_data[col] = 0

    # Predict price
    suggested_price = model.predict(input_data)
    return suggested_price[0]
```
